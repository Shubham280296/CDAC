## 1. Import Statements

```python
###-----------------
### Import Libraries
###-----------------

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

## 2. Setup Global Parameters

```python
###----------------
### Some parameters
###----------------

# Directory locations
inpDir = '../../Deep Neural Network\Classwork\input' # Set path of input directory as per your system 

RANDOM_STATE = 24 # REMEMBER: to remove at the time of promotion to production
np.random.seed(RANDOM_STATE)
rng = np.random.default_rng(seed = RANDOM_STATE) # Set Random Seed for reproducible  results

NOISE = 0.2
EPOCHS = 200  # number of epochs
ALPHA = 0.1  # learning rate
N_SAMPLES = 1000

# parameters for Matplotlib
params = {'legend.fontsize': 'medium',
          'figure.figsize': (15, 6),
          'axes.labelsize': 'large',
          'axes.titlesize':'large',
          'xtick.labelsize':'large',
          'ytick.labelsize':'large'
         }

plt.rcParams.update(params)

CMAP = plt.cm.coolwarm
plt.style.use('seaborn-v0_8-darkgrid') # plt.style.use('ggplot')
```

```python
df = pd.read_csv(os.path.join(inpDir,'breast_cancer_mod.csv')) ### read dataset
df
```

```python
df['target'].value_counts() ## unique target values
```

```python
df.info()
```

```python
df.describe()
```

```python
### checking whether dataset is balanced or not
fig,ax = plt.subplots()
df['target'].value_counts().plot(kind='bar')
```

```python
test_size = 0.2
features = df.drop('target',axis=1)
labels = df['target']

X_train,X_test, y_train, y_test = train_test_split(features,labels,test_size=test_size,random_state=RANDOM_STATE) ### train test split

X_train['total'] = X_train.sum(axis=1)  ### counts total of all columns
X_train['target'] = y_train
X_train.sort_values('total',inplace=True) ### sorting on total
X_train
```

```python
### visualize total vs target
plt.scatter(X_train['total'],X_train['target'])
```

```python
### plot total vs no of rows with total
tmp_df = X_train.groupby(['total','target']).agg({'target':'count'})
tmp_df.columns=['count']
tmp_df.reset_index(inplace=True)
tmp_df.pivot(index='total',columns='target',values='count').plot(kind='bar');
```

```python
tmp_df
```

```python
accuracy = {}
for b in range(X_train.shape[1]):
    X_train['pred']=X_train['total']>=b ### b --> threshold (or boundary)
    accuracy[b]=accuracy_score(X_train['target'],X_train['pred'])

b_best = max(accuracy,key=accuracy.get) ### best threshold
b_best
```

```python
accuracy
```

```python
X_test['total'] = X_test.sum(axis=1) ### counts total of all columns
X_test['target'] = y_test
X_test.sort_values('total',inplace=True) ### sorting on total
X_test
```

```python
### visualize total vs target for test data
plt.scatter(X_test['total'],X_test['target'])
```

```python
### plot total vs no of rows with total
tmp_tst = X_test.groupby(['total','target']).agg({'target':'count'})
tmp_tst.columns=['count']
tmp_tst.reset_index(inplace=True)
tmp_tst.pivot(index='total',columns='target',values='count').plot(kind='bar');
```

```python
X_test['pred']=X_test['total']>=b_best ### threshold = best threshold for training data
accuracy=accuracy_score(X_test['target'],X_test['pred'])
accuracy
```
