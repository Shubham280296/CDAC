```python
import numpy as np
from matplotlib import pyplot as plt
v = np.array([2,4,7])
u = np.array([1,2,3])
```

```python
u.dot(v)
```

```python
np.dot(u,v)
```

```python
y = np.zeros(101)
x = np.arange(101)
```

```python
for i in range(0,101):
  a = np.linalg.norm(u,ord =i)
  y[i] = a

```

```python
y = np.array(y)
```

```python
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```
