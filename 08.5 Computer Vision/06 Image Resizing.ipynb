### Resize using size

```python
import cv2

img = cv2.imread('images/lenna.png')

## Resize the image to a fixed size
img_resized = cv2.resize(img,(100,200))

cv2.imshow("Original image", img)
cv2.imshow("Resized image", img_resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_resized.shape
```

### Resize using scalling factor

```python
import cv2

img = cv2.imread('images/lenna.png')

## Resize the image to a fixed size
img_resized = cv2.resize(img,None, fx = 0.5, fy=0.75)

cv2.imshow("Original image", img)
cv2.imshow("Resized image", img_resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_resized.shape
```

### Resize using scikit-image

```python
from skimage.transform import resize
import matplotlib.pyplot as plt

image = plt.imread('images/lenna.png')

## RESIZE BILINEAR INTERPOLATION
resized_img = resize(image, (1000,1200), anti_aliasing= True)

## RESIZE WITH BICUBIC INTERPOLATION
resized_img = resize(image, (1300,1300), anti_aliasing= True, order = 3)

plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(resized_img)
```

```python
import cv2

img = cv2.imread('images/lenna.png')

## Horizontal flip
horizontal_flip = cv2.flip(img,0)

## Verical flip
vertical_flip = cv2.flip(img,1)

## Diagonal flip
diagonal_flip = cv2.flip(img,-1)

cv2.imshow("Original image", img)
cv2.imshow("Horizontal flip image", horizontal_flip)
cv2.imshow("Verical flip image", vertical_flip)
cv2.imshow("Diagonal flip image", diagonal_flip)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
