1. Size and shape of the image        

```python
import cv2
import matplotlib.pyplot as plt

# Load the image using OpenCV
image = cv2.imread("sachin.jpg")

# Extract image properties - size and shape
print("Size of the image :", image.size)
print("Shape of the image :", image.shape)
```

```python
# Determine color space model
color_space = "Unknown"
if image.shape[2] == 1:
    color_space = "Grayscale"
elif image.shape[2] == 3:
    color_space = "BGR"  # OpenCV uses BGR by default
print("The color space model :", color_space)
```

### 4. Download the video nature.mp4 

1. Find height, width, total frames and fps of the video        

```python
import cv2
vid = cv2.VideoCapture('nature.mp4')

width = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(vid.get(cv2.CAP_PROP_FPS))
frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))

print("Video Resolution: {} x {}".format(height, width))
print("Total Frames:", frames)
print("FPS:", fps)
print(f"Duration: {round(frames/fps)} seconds")
```

2. Play the video in frames

```python
import cv2
vid = cv2.VideoCapture('nature.mp4')
while True:
    ret, frame = vid.read()
    if not ret:
        break
    cv2.imshow('Video', frame)
    if cv2.waitKey(5)== ord('q'):
        break
cv2.destroyAllWindows()
```

3. Save the image at 50th second.

```python
# get frames with specific timestamp
timestamp = '00:00:50.00'
fps = int(vid.get(cv2.CAP_PROP_FPS))
frames = int(vid.get(cv2.CAP_PROP_FRAME_COUNT))
hours , minutes, seconds = [float(i) for i in timestamp.split(':')]

trigger_frame = (((hours*60)+minutes)*60+seconds)*fps
print(frames, trigger_frame)
```

```python
# Go to frame selected by timestamp
vid.set(1, trigger_frame)
success, frame = vid.read()

# save the frame into an image file
if success:
    cv2.imwrite("specific_timestamp_frame.jpg", frame)
```
