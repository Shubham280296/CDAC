### 1. Apply following transformations on the image.

Image to use : sachin.jpg

Location: images folder on https://mitu.co.in/dataset

1. Translate the image towards right by 400 and up by 150 pixels.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('sachin.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

## Transformatiion matrix
T = np.float32([[1,0,400],[0,1,-150]])

## We use warpAffine to transform the image using matrix ,T
img_translation = cv2.warpAffine(img, T, (img.shape[1], img.shape[0]))

plt.imshow(img_translation)  # Convert BGR to RGB for matplotlib
plt.show()
```

2. Rotate the image 60 degrees anti clockwise.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('sachin.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

center = (img.shape[0]/2,img.shape[1]/2)
angle = 60

## Create rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle,1.0)

## We use warpAffine to transform the image using matrix ,T
img_rotated = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))

plt.imshow(img_rotated)
```

3. Resize the image to 75% of original dimensions.

```python
import cv2

img = cv2.imread('sachin.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

## Resize the image to a fixed size
img_resized = cv2.resize(img,None, fx = 0.75, fy=0.75)

plt.imshow(img_resized)
```

4. Flip the image horizontally, diagonally and vertically. Show all four images in the row.

```python
import cv2

img = cv2.imread('sachin.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_hori = cv2.flip(img, 1)
img_dia = cv2.flip(img, -1)
img_vert = cv2.flip(img, 0) 

plt.figure(figsize=(16,4))

plt.subplot(141)
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')

plt.subplot(142)
plt.title("Horizontally flipped Image")
plt.imshow(img_hori)
plt.axis('off')

plt.subplot(143)
plt.title("Diagonally flipped Image")
plt.imshow(img_dia)
plt.axis('off')

plt.subplot(144)
plt.title("Vertically flipped Image")
plt.imshow(img_vert)
plt.axis('off')

plt.show();
```

5. Crop the image at, 500,300 to 560, 400 and save in another file

```python
import cv2

img = cv2.imread('sachin.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

## Resize the image to a fixed size
img_cropped = img[300:400, 500:560,:]

plt.imshow(img_cropped)

img_cropped = cv2.cvtColor(img_cropped, cv2.COLOR_RGB2BGR)
cv2.imwrite('Cropped.png', img_cropped)
```
