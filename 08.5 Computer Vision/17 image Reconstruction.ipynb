```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Datasets/cat_damaged.png')
height, width = image.shape[:-1]
height, width
```

```python
cv2.imshow('Damaged cat', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask = np.zeros([height, width], dtype = np.uint8)
```

```python
## Converting all pixels greater than zero to black whileblack becomes white
for i in range(height):
    for j in range(width):
        if image[i,j].sum()==0:
            mask[i,j] = 255
```

```python
plt.subplot(121)
plt.imshow(image)
plt.subplot(122)
plt.imshow(mask, cmap='gray')
```

```python
## Inpaint
dst = cv2.inpaint(image, mask, 3, cv2.INPAINT_NS)
cv2.imshow('Reconstructed cat', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
