### Using default rotation

```python
img = cv2.imread('images/lenna.png', 0)

## Rotate the image 90 degrees anticlockwise
img_rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)

cv2.imshow("Original image", img)
cv2.imshow("Rotated image", img_rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Using rotation matrix

```python
img = cv2.imread('images/lenna.png', 0)

## Find the center point and angle of rotation
## center = (img.shape[1]//2,img.shape[0]//2)
center = (250,250)
angle = -60

## Create rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle,1.0)

## We use warpAffine to transform the image using matrix ,T
img_rotated = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))

cv2.imshow("Original image", img)
cv2.imshow("Rotated image", img_rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
