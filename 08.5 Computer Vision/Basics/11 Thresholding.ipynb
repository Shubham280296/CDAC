### Binary thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

## Load image
img = cv2.imread('images/lenna.png',0)

## Filtering pixels with threshold
binary_image = img > 127
```

```python
binary_image
```

```python
forfront = binary_image.astype(np.uint8)*255
```

```python
## Show original and smoothed image
cv2.imshow("Original image", img)
cv2.imshow("Thesholded  image", forfront)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## Show original and smoothed image
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.imshow(img, cmap = 'gray')
plt.title("Original image")
plt.subplot(122)
plt.imshow(binary_image, cmap = 'gray')
plt.title("Binary image")
```

### Inverse Binary thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

## Load image
img = cv2.imread('images/lenna.png',0)

## Filtering pixels with threshold
binary_image = img < 127

## Show original and smoothed image
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.imshow(img, cmap = 'gray')
plt.title("Original image")
plt.subplot(122)
plt.imshow(binary_image, cmap = 'gray')
plt.title("Inverse Binary image")
```

#### With in-built functions

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

## Load image
img = cv2.imread('images/lenna.png',0)

## Thresholding
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)

## Show original and smoothed image
plt.figure(figsize = (15,5))
plt.subplot(131)
plt.imshow(img, cmap = 'gray')
plt.title("Original image")
plt.subplot(132)
plt.imshow(thresh1, cmap = 'gray')
plt.title("Binary image")
plt.subplot(133)
plt.imshow(thresh2, cmap = 'gray')
plt.title("Inverse Binary image")
```

### Otsu thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

## Load image
img = cv2.imread('images/lenna.png',0)

## Thresholding
value, thresh_otsu = cv2.threshold(img, 120, 255, cv2.THRESH_OTSU)

## Threshold
print("Value of threshold = ", value)

## Show original and smoothed image
plt.figure(figsize = (15,5))
plt.subplot(131)
plt.imshow(img, cmap = 'gray')
plt.title("Original image")
plt.subplot(132)
plt.imshow(thresh_otsu, cmap = 'gray')
plt.title("Binary image")
plt.show();
```
