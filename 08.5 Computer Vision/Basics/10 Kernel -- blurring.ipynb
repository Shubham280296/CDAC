## Averaging Kernel -- blurred image

```python
import cv2
import numpy as np

## Load image
img = cv2.imread('images/lenna.png')

## Define a 5x5 averaging kernel
kernel = np.ones((5,5), np.float32)/25

## Apply a kernel to smooth the image
smoothed_image = cv2.filter2D(img, -1, kernel)

## Stacking to view both at once
new_image = np.hstack([img, smoothed_image])

## Show original and smoothed image
cv2.imshow("Original image and Smoothed image", new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Gussian Kernel -- blurred image

```python
import cv2
import numpy as np

## Load image
img = cv2.imread('images/lenna.png')

## Define kernel
kernel_size = 15
sigma = 0.35*((kernel_size-1)*0.5-1)+0.9
kernel = cv2.getGaussianKernel(kernel_size, sigma)

## Apply a kernel to smooth the image
smoothed_image = cv2.filter2D(img, -1, kernel)

## Stacking to view both at once
new_image = np.hstack([img, smoothed_image])

## Show original and smoothed image
cv2.imshow("Original image and Smoothed image", new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Median Kernel -- blurred image

```python
import cv2
import numpy as np

## Load image
gray = cv2.imread('images/lenna.png', 0)

## Define kernel
kernel_size = 11
median = cv2.medianBlur(gray, kernel_size)

plt.figure(figsize = (10,5))
plt.subplot(121)
plt.imshow(gray, cmap = 'gray')
plt.title("Original image")
plt.subplot(122)
plt.imshow(median, cmap = 'gray')
plt.title("Median filtered image")
```

## Sharpening Kernel

```python
import cv2
import numpy as np

## Load image
img = cv2.imread('images/lenna.png')

## Define a 5x5 averaging kernel
kernel = np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])

## Apply a kernel to smooth the image
smoothed_image = cv2.filter2D(img, -1, kernel)

## Show original and smoothed image
cv2.imshow("Original image", img)
cv2.imshow("Smoothed image", smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Slective blurring

```python
import cv2
import numpy as np

## Load image
img = cv2.imread('images/lenna.png')

## Choose only part to be blurred
img_cropped = img[100:300, 100:300,:]

## Define kernel and apply on cropped image
kernel_size = 15
blurred_img = cv2.medianBlur(img_cropped, kernel_size)

## Replace with blurred part
img_cropped = img[100:300, 100:300,:] = blurred_img

cv2.imshow("Smoothed image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
