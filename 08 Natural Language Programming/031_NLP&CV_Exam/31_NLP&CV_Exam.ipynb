## Q1

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Load a sample image
img = cv2.imread('sample.jpg')
img.shape
```

```python
## Resize the image to half its original size
img_resize = cv2.resize(img, None, fx = 0.5, fy = 0.5) ## Resize scale = 0.5 along both axis
print("Size of image after resizing: ",img_resize.shape)

## Display original and transformed images
cv2.imshow("Original Image", img)
cv2.imshow("Resized Image", img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## Rotate it by 45 degrees clockwise
center = [int(img.shape[0]/2), int(img.shape[1]/2)]  ## rotate about center
angle = -45

T = cv2.getRotationMatrix2D(center, angle, 1)

img_rotated = cv2.warpAffine(img, T, (img.shape[1], img.shape[0]))

## Display original and transformed images
cv2.imshow("Original Image", img)
cv2.imshow("Rotated Image", img_rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##Save Rotated and resized images
cv2.imwrite("Rotated_image.png", img_rotated)

cv2.imwrite("Resized_image.png", img_resize)
```

```python
## Load a sample image
img = cv2.imread('sample.jpg', 0)

### Sobel edge detector
sobelx = cv2.Sobel(img, cv2.CV_64F, 1,0, ksize=3)  
sobely = cv2.Sobel(img, cv2.CV_64F, 0,1, ksize=3)  

##Cobine sobel images
combined_sobel = cv2.add(sobelx, sobely)

value, thresh1 = cv2.threshold(combined_sobel, 90, 255, cv2.THRESH_BINARY)

```

```python
## Canny Edge detector
import cv2
import numpy as np

## Load image
img = cv2.imread('sample.jpg', 0)

edges = cv2.Canny(img, 50, 200)
```

```python
## Display original and edge-detected image
plt.figure(figsize=(18,6))
plt.subplot(131)
plt.imshow(img, cmap = 'gray')
plt.title("Original image")
plt.subplot(132)
plt.imshow(thresh1, cmap = 'gray')
plt.title("Both edge (sobel x, sobel y) image")
plt.subplot(133)
plt.imshow(edges, cmap = 'gray')
plt.title("horizontal edge (sobel x) image")
```

## Analysis

## Build model

```python
model = Sequential()

## Layer 1
model.add(Conv2D(filters = 16,
                 kernel_size = (3,3),
                 input_shape = (64,64,3),
                 activation = 'relu')) ##62*62*16
model.add(MaxPool2D(pool_size = (2,2))) ##31*31*16

## Layer 2
model.add(Conv2D(filters = 8,
                 kernel_size = (3,3),
                 activation = 'relu')) ##29*29*8
model.add(MaxPool2D(pool_size = (2,2))) ##14*14*8

model.add(Flatten())

## Fully connected layers
model.add(Dense(160, activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.summary()
```

## Compile and fit model

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = 'accuracy')

history = model.fit(train, epochs = 10, validation_data =test)
```

## Visualize results

```python
hist = pd.DataFrame(history.history)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(121)
plt.plot(hist["loss"], label = 'loss')
plt.plot(hist["val_loss"], label = 'val_loss')
plt.legend()
plt.title('Loss vs epoch')
plt.subplot(122)
plt.plot(hist["accuracy"], label = 'accuracy')
plt.plot(hist["val_accuracy"], label = 'val_accuracy')
plt.legend()
plt.title('Accuracy vs epoch')
```
