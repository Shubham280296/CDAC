```python
aplhabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
int_to_char = dict(enumerate(aplhabets))
int_to_char
```

```python
char_to_int = dict((v,i) for i,v in enumerate(aplhabets))
char_to_int
```

# Seq_len = 1

## Data Preparation

```python
seq_length = 1
x = []
y = []
for i in range(len(aplhabets)-seq_length):
    seq_in = [aplhabets[i]]
    seq_out = aplhabets[i+1]
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]  
    print(seq_in, '->', seq_out)
```

## Prepare the output variable

```python
from keras.utils import to_categorical
```

```python
y_new = to_categorical(y)
y_new
```

## Prepare input variables

```python
import numpy as np
```

```python
x = np.reshape(x,(25,1,1))
x.shape
```

```python
x_new = x/25
x_new.shape
```

## Creat model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, SimpleRNN
```

## LSTM

```python
model = Sequential()
model.add(Dense(25,input_shape=(1,1), activation='relu'))
model.add(LSTM(32))
model.add(Dense(26, activation='softmax'))
```

```python
## Compile model
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics = 'accuracy')

## Train model 
model.fit(x_new, y_new, epochs = 1000, batch_size=1)
```

## Predict

```python
model.evaluate(x_new, y_new, batch_size=1)
```

```python
## Predict the characterin sequence
new = 'R'
valid = np.reshape(char_to_int[new],(1,1))/25
print(int_to_char[np.argmax(model.predict(valid))])
```

# Seq_len = 3 (>1).

## Data preparation

```python
seq_length = 3
x = []
y = []
for i in range(len(aplhabets)-seq_length):
    seq_in = aplhabets[i:i+seq_length]
    seq_out = aplhabets[i+seq_length]
    x.append([char_to_int[char] for char in seq_in])
    y += [char_to_int[seq_out]]  
    print(seq_in, '->', seq_out)
```

```python
x = np.reshape(x,(23,seq_length,1))
x.shape
```

```python
x_new = x/25
x_new
```

```python
y_new = to_categorical(y)
y_new
```

## Build model

```python
model = Sequential()
model.add(Dense(25,input_shape=(3,1), activation='relu'))
model.add(LSTM(32))
model.add(Dense(26, activation='softmax'))
```

```python
## Compile model
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics = 'accuracy')

## Train model 
model.fit(x_new, y_new, epochs = 1000, batch_size=1)
```

## Predict

```python
model.evaluate(x_new, y_new, batch_size=1)
```

```python
## Predict the characterin sequence
new = 'HIJ'
valid = np.reshape([char_to_int[char] for char in new],(1,3))/25
print(int_to_char[np.argmax(model.predict(valid))])
```

## Simple RNN

```python
model = Sequential()
model.add(Dense(25,input_shape=(3,1), activation='relu'))
model.add(SimpleRNN(32))
model.add(Dense(26, activation='softmax'))
```

```python
## Compile model
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics = 'accuracy')

## Train model 
model.fit(x_new, y_new, epochs = 1000, batch_size=1)
```

```python
## Predict the characterin sequence
new = 'EFG'
valid = np.reshape([char_to_int[char] for char in new],(1,3))/25
print(int_to_char[np.argmax(model.predict(valid))])
```
