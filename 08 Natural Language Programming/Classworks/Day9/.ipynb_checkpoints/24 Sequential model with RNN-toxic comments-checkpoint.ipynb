```python
training_sentences = []
training_labels = []
testing_sentences = []
testing_labels = []

for s,l in train_data:
    training_sentences.append(str(s.numpy()))
    training_labels.append(l.numpy())
    
for s,l in test_data:
    testing_sentences.append(str(s.numpy()))
    testing_labels.append(l.numpy())

len(training_labels),len(testing_labels)
```

```python
training_labels_final = np.array(training_labels)
testing_labels_final = np.array(testing_labels)
training_labels_final.size, testing_labels_final.size
```

```python
vocab_size = 10000
embedding_dim = 16
max_length = 120
trunc_type = 'post'
```

```python
tokenizer = Tokenizer(num_words = vocab_size)
tokenizer.fit_on_texts(training_sentences)
word_index = tokenizer.word_index
len(word_index)
```

```python
word_index
```

```python
sequences = tokenizer.texts_to_sequences(training_sentences)
padded = pad_sequences(sequences, maxlen = max_length, truncating = trunc_type)

testing_sequences = tokenizer.texts_to_sequences(testing_sentences)
testing_padded = pad_sequences(testing_sequences, maxlen = max_length, truncating = trunc_type)

len(sequences), len(sequences[0]), len(padded[0]), padded.shape, testing_padded.shape, padded.dtype
```

## RNN - Build Model 

```python
model_rnn = Sequential()
model_rnn.add(Embedding(vocab_size, embedding_dim, input_length = max_length))
model_rnn.add(SimpleRNN(32))
model_rnn.add(Dense(10, activation ='relu'))  
model_rnn.add(Dense(1, activation ='sigmoid'))   

model_rnn.summary()
```

```python
model_rnn.compile(loss = 'binary_crossentropy', optimizer='adam', metrics = 'accuracy')

history_rnn = model_rnn.fit(padded,training_labels_final, epochs=10, validation_data = (testing_padded,testing_labels_final))
```

```python
hist_rnn = pd.DataFrame(history_rnn.history)
fig = plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.grid()
plt.title("Loss - Simple RNN")
plt.xlabel('Epoch')
plt.plot(hist_rnn['loss'], label = 'loss', marker='.')
plt.plot(hist_rnn['val_loss'], label = 'val_loss', marker='.')
plt.legend()

plt.subplot(1,2,2)
plt.grid()
plt.title("Accuracy - Simple RNN")
plt.xlabel('Epoch')
plt.plot(hist_rnn['accuracy'], label = 'accuracy', marker='.')
plt.plot(hist_rnn['val_accuracy'], label = 'val_accuracy', marker='.')
plt.legend()
```

## Bidirectional LSTM - Build Model 

```python
model_lstm = Sequential()
model_lstm.add(Embedding(vocab_size, embedding_dim, input_length = max_length))
model_lstm.add(Bidirectional(LSTM(32)))
model_lstm.add(Dense(10, activation ='relu'))  
model_lstm.add(Dense(1, activation ='sigmoid'))   

model_lstm.summary()
```

```python
model_lstm.compile(loss = 'binary_crossentropy', optimizer='adam', metrics = 'accuracy')

history_lstm = model_lstm.fit(padded,training_labels_final, epochs=10, validation_data = (testing_padded,testing_labels_final))
```

```python
hist_lstm = pd.DataFrame(history_lstm.history)
fig = plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.grid()
plt.title("Loss - Bidirectional LSTM")
plt.xlabel('Epoch')
plt.plot(hist_lstm['loss'], label = 'loss', marker='.')
plt.plot(hist_lstm['val_loss'], label = 'val_loss', marker='.')
plt.legend()

plt.subplot(1,2,2)
plt.grid()
plt.title("Accuracy - Bidirectional LSTM")
plt.xlabel('Epoch')
plt.plot(hist_lstm['accuracy'], label = 'accuracy', marker='.')
plt.plot(hist_lstm['val_accuracy'], label = 'val_accuracy', marker='.')
plt.legend()
```

## Bidirectional GRU - Build Model 

```python
model_gru = Sequential()
model_gru.add(Embedding(vocab_size, embedding_dim, input_length = max_length))
model_gru.add(Bidirectional(GRU(32)))
model_gru.add(Dense(10, activation ='relu'))  
model_gru.add(Dense(1, activation ='sigmoid'))   

model_gru.summary()
```

## Validation

```python
new1 = 'It was worst movie. Please do not watch it. Waste of money and time.'
new2 = 'A must watch movie. I loved watching it. It was just amazing!'
```

```python
def predict(model,text):
    validation_sequences = tokenizer.texts_to_sequences([text])
    validation_padded = pad_sequences(validation_sequences, maxlen = max_length, truncating = trunc_type)

    y_pred = model.predict(validation_padded)
    if y_pred[0,0]<0.5:
        return((1-y_pred[0,0]), 'Negative')
    else:
        return((y_pred[0,0]), 'Positive')
```

```python
predict(model_rnn,new1), predict(model_rnn,new2)
```

```python
predict(model_lstm,new2), predict(model_lstm,new1)
```

```python
predict(model_gru,new1), predict(model_gru,new2)
```
