## Import the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import *
from keras.layers import *
from keras.datasets import imdb
from keras.utils import pad_sequences
```

### Define the transformer block

```python
class TransformerBlock(Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate = 0.1):
        ## embed_dim : dimentionality of ip and op
        ## num_heads : controls no of attention heads
        ## ff_dim : dimentionality of feed forward neural network
        
        super().__init__()
        self.att = MultiHeadAttention(num_heads = num_heads, key_dim = embed_dim)
        ## Creats MultiHeadAttention layer responsible for learning long range dependencies
        self.ffn = Sequential([
            Dense(ff_dim, activation = 'relu'),
            Dense(embed_dim)
        ])
        ## creates a feed forward neural network
        
        self.layernorm1 = LayerNormalization(epsilon =1e-6)
        self.layernorm2 = LayerNormalization(epsilon =1e-6)  
        ## Layer normalization
        
        self.dropout1 = Dropout(rate)
        self.dropout2 = Dropout(rate)   
        ## Create dropout layers
        
    def call(self, inputs, training):
        attn_output= self.att(inputs, inputs)
        ## applies 
        
        attn_output = self.dropout1(attn_output, training = training)
        
        
        out1 = self.layernorm1(inputs+attn_output)
        
        
        
        ffn_output = self.ffn(out1)
        ## Passes normalized op though ffed forward network
        
        ffn_output = self.dropout2(ffn_output, training = training)
        ## dropout to feedforward op
        
        return self.layernorm2(out1+ffn_output)
        ## adds feedforward op to previous layer op and applies final normalization layer
```

```python
class TokenAndPositionEmbedding(Layer):
    def __init__(self, maxlen, vocab_size, embed_dim):
        ## maxlen : max len of ip sequences the model will handle
        ## vocab_size : total no of unique tokens in vocab
        ## embed_dim : dimensionality of embedding
                ## how each token and its position will be represented as a vector
        
        super().__init__()
        self.token_emb = Embedding(input_dim = vocab_size, output_dim = embed_dim)
        ## Embedding layer that maps each token in ip sentence to a dense vector of size embed_dim
        self.pos_emb = Embedding(input_dim = maxlen, output_dim = embed_dim)
        ## Embedding layer that maps each position in ip sentence from 0 to maxlen-1 to a dense vector of size embed_dim 
        
    def call(self, x):
        maxlen = tf.shape(x)[-1]
        ## Extracts actual length of current ip sequence
        
        positions = tf.range(start = 0, limit = maxlen, delta = 1)
        ## tensor of positions from 0 to maxlen-1
        
        positions = self.pos_emb(positions)
        ## looks up position embeddings for each position in sequence
        
        x = self.token_emb(x)
        ## looks up token embeddings for each token in sequence
        
        return x + positions
        ## adds token and position embeddings element-wise 
```

```python
vocab_size = 20000  ## only consider first 20k unique words
maxlen = 200 ## only consider first 200 words of each movie review
```

```python
(x_train, y_train),(x_val, y_val) = imdb.load_data(num_words = vocab_size)
print(len(x_train), "Training sequences")
print(len(x_val),"Validation sequences")
```

```python
x_train = pad_sequences(x_train, maxlen=maxlen)
x_val = pad_sequences(x_val, maxlen=maxlen)
x_train.shape, x_val.shape
```

## Combine all together

```python
embed_dim = 32 ## Embedding size for each token
num_heads = 2 ## npo of attention heads
ff_dim = 32 ## Hidden layer size in feed forward network inside transformer

inputs = Input(shape=(maxlen,))
embedding_layer = TokenAndPositionEmbedding(maxlen, vocab_size, embed_dim)
x = embedding_layer(inputs)
transformer_block = TransformerBlock(embed_dim, num_heads, ff_dim)
x = transformer_block(x)
x = GlobalAveragePooling1D()(x)
x = Dropout(0.1)(x)
x = Dense(20,activation = 'relu')(x)
x = Dropout(0.1)(x)
outputs =  Dense(2,activation = 'softmax')(x)

model = Model(inputs = inputs, outputs = outputs)
model.summary()
```

### Model compliation and fitting

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer ='adam', metrics = 'accuracy')

history = model.fit(x_train, y_train, batch_size= 32, epochs = 10, validation_data = (x_val, y_val))
```

```python
hist = pd.DataFrame(history.history)
fig = plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.grid()
plt.title("Loss - Simple RNN")
plt.xlabel('Epoch')
plt.plot(hist['loss'], label = 'loss', marker='.')
plt.plot(hist['val_loss'], label = 'val_loss', marker='.')
plt.legend()

plt.subplot(1,2,2)
plt.grid()
plt.title("Accuracy - Simple RNN")
plt.xlabel('Epoch')
plt.plot(hist['accuracy'], label = 'accuracy', marker='.')
plt.plot(hist['val_accuracy'], label = 'val_accuracy', marker='.')
plt.legend()
```
