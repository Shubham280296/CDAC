### Q1 Open the wikipedia page https://en.wikipedia.org/wiki/Mount_Everest

```python
## Opening url
url = urlopen('https://en.wikipedia.org/wiki/Mount_Everest')
## Reading url
data = url.read()
print(data)
```

```python
## Parsing and getting only text
soup = BeautifulSoup(data, 'html.parser')
text = soup.get_text(strip = True)
print(text)
```

```python
## Create tokens
tokens = word_tokenize(text)
tokens
```

1. Find the count of Total “Mount” in the text.

```python
count = tokens.count("Mount")
count
```

2. Save the extracted plain text contents in file mount.txt in upper case.

```python
file_object = open('mount.txt', 'w')
file_object.write(text.upper())
file_object.close(  )
```

```python
import requests
from docx import Document

url = 'https://icml.cc/Conferences/2013/wp-content/uploads/2013/06/Machine-Learning-and-Natural-Language-Processing.docx'
response = requests.get(url)

with open('mlnlp.docx', 'wb') as f:
  f.write(response.content)

document = Document('mlnlp.docx')
```

Read and display first 5 paragraphs from it

```python
text = []
for paragraph in document.paragraphs[:5]:
  text += paragraph.text
  text = "".join(text)
print(text)
```

Count total ‘The’ present in the text.

```python
count = text.count('The')
print(f"\nThe word 'The' appears {count} times in the text.")
```
