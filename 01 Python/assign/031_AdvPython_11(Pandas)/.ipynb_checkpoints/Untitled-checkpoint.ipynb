```python
#Q1
import pandas as pd
df1 = pd.read_csv("movies11.csv")
df2 = pd.read_csv("movie12.csv",header=None)
df3 = pd.read_csv("movies13.csv",header=None)

df4 = df2.set_axis(['movieId', 'title', 'genres'], axis=1)
df5 = df3.set_axis(['movieId', 'title', 'genres'], axis=1)
dataset = [df1,df4,df5]
df6 = pd.concat(dataset)
df6.set_index('movieId',inplace = True)
```

```python
#Q2
masala = ["Action","Romance","Comedy","Thriller"]
for i,j in zip(df6['genres'],df6['title']):
    str1=str(i).split("|")
    for k in masala:
        if not(k in str1):
            break
    else:
        print(j)
```

```python
#Q3
genre = df6['genres'].str.split("|")
value = []
for i in genre:
    value += i
unique = list(set(value))
counter = []
for i in unique:
    counter.append(value.count(i))

import matplotlib.pyplot as plt
plt.pie(counter,labels=unique,startangle=90)
plt.title('No of movies in each genre')
plt.show()
```

```python
#Q4
df = pd.read_csv("rating11.csv")
g=df.groupby('movieId').mean()
h= pd.merge(df6, g['rating'], on='movieId',how = 'left')
p = h.drop(['rating'],axis=1)
p
```

```python
#Q5
data = h.dropna()
rate = []
gen = []
for i,j in zip(data['genres'],data['rating']):
    str1=str(i).split("|")
    for k in str1:
        rate.append(j)
        gen.append(k)
new = pd.DataFrame(
    {'genres': gen,
     'rating': rate,
    })
rategen=new.groupby('genres').mean()
```

```python
import matplotlib.pyplot as plt

slices=list(rategen['rating'])
activities=list(rategen.index)
plt.pie(slices,labels=activities,startangle=90,autopct='%1.1f%%')
plt.title('No of students in each location')
plt.show()
```

```python
#Q5
label = [0,1,2,3,4,5]
rating = list(data['rating'].astype('int32'))
counter = []
for i in label:
    counter.append(rating.count(i))
plt.bar(label, counter)
plt.xlabel('average rating', fontweight ='bold', fontsize = 15) 
plt.ylabel('No of movies', fontweight ='bold', fontsize = 15) 
plt.show()
```
