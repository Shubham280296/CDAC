```python
import pandas as pd
df1 = pd.read_excel("student.xlsx")
df2 = pd.read_excel("student2.xlsx",header=None)
df3 = pd.read_excel("interviewer_rating.xlsx")
```

```python
df4 = df2.set_axis(['stuid', 'name', 'marks'], axis=1)
dataset = [df1,df4]
df5 = pd.concat(dataset)
```

```python
g=df3.groupby("studentid").mean()
h=df5.merge(g['rating'], left_on='stuid', right_on='studentid')
```

```python
def Technology(num1,num2):
    if num1>80 and num2>4:
        return "Python project"
    elif num1>90 and num2>3:
        return "Java project"
    elif num2>4:
        return "Python training"
    else:
        return "Java training"
h['technology']=h.apply(lambda x: Technology(x['marks'], x['rating']), axis=1)
```

```python
def location(num1):
    if num1>80:
        return "Pune"
    elif num1>=70 :
        return "Delhi"
    else:
        return "Mumbai"
h['location']=h.apply(lambda x: location(x['marks']), axis=1)
```

```python
import matplotlib.pyplot as plt

slices=list(pd.value_counts(h['location']))
activities=list(pd.unique(h['location']))
cols=['g','c','m']
plt.pie(slices,labels=activities,colors=cols,startangle=90)

plt.title('No of students in each location')
plt.show()
```
