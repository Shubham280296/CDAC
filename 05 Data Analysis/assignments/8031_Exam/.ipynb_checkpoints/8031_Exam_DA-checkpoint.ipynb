```python
#Q2
```

```python
#load dataset again
titanic = pd.read_csv("Titanic-Dataset.csv")
titanic.head(5)
```

```python
# Compute mean of Fare column
mean = titanic["Fare"].mean()
mean
```

```python
#replace Nan values in Fare column with mean
titanic["Fare"].fillna(mean,inplace=True)
# Check if Nan values are replaced 
titanic["Fare"].isna().sum()
```

```python
# Calculate 0.25 quartile, 0.75 quartile and standard deviation
q1 = titanic['Fare'].quantile(0.25)
q3 = titanic['Fare'].quantile(0.75)
iqr = q3-q1
std = titanic['Fare'].std()

# Define Lower and Upper bound
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

# Check values lower than lower bound and higher than upper bound to find outliers
outliers = titanic[(titanic["Fare"]<lower)|(titanic["Fare"]>upper)]
outliers.head()
```

```python
# Boxplot for Fare values
plt.figure(figsize=(12,8))
sns.boxplot(titanic["Fare"])
plt.axhline(q1,color = "g",lw = 0.5,label = 'q1')
plt.axhline(q3,color = "g",lw = 0.5,label = 'q3')
plt.axhline(lower,color = "red",lw = 0.5,label = 'lower bound')
plt.axhline(upper,color = "red",lw = 0.5,label = 'upper bound')
plt.legend()
```

```python
#load dataset again
titanic = pd.read_csv("Titanic-Dataset.csv")
titanic.head(5)
```

```python
# Column with max missing values
col = titanic.isna().sum().idxmax()
print("Column with max missing values - ", col)
```

```python
# Means of the Pclass values
mean1 = (titanic["Pclass"] == 1).mean()
mean2 = (titanic["Pclass"] == 2).mean()
mean3 = (titanic["Pclass"] == 3).mean()
```

```python
#Dictionary to map 'Pclass' to mean
mean_dict = {1: mean1, 2: mean2, 3: mean3}

# Replace missing values in 'max_missing_data' with mean of respective 'Pclass'
titanic[col] = titanic.apply(lambda row: mean_dict[row['Pclass']] if pd.isnull(row[col]) else row[col],axis=1)
titanic.head()
```

```python
#load dataset again
titanic = pd.read_csv("Titanic-Dataset.csv")
titanic.head(5)
```

```python
# Filter non survived male
titanic_filter = titanic[(titanic["Survived"]==0)&(titanic["Sex"]=="male")]
titanic_filter.head(5)
```

```python
# Calculate mean,Median,Mode,Quartile Ranges, Standard deviation of age for "Not Survived Male"
mean = titanic_filter["Age"].mean()
print("mean age for Not Survived Male - ",mean)
median = titanic_filter["Age"].median()
print("median age for Not Survived Male - ",median)
mode = titanic_filter["Age"].mode()[0]
print("mode age for Not Survived Male - ",mode)
q1 = titanic_filter["Age"].quantile(0.25)
print("25% quartile age for Not Survived Male - ",q1)
q3 = titanic_filter["Age"].quantile(0.75)
print("75% quartile age for Not Survived Male - ",q3)
iqr = q3-q1
print("Inter quartile range of age for Not Survived Male - ",iqr)
std = titanic_filter["Age"].std()
print("standard deviation age for Not Survived Male - ",std)
```
