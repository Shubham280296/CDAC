```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("alphabet_stock_data.csv")
df.head(20)
```

1. Write a Pandas program to create a line plot of the historical stock prices of 
Alphabet Inc. between two specific dates.

```python
start_date = pd.to_datetime("2020-04-07")
end_date = pd.to_datetime("2020-09-24")
df["Date"]=pd.to_datetime(df["Date"])
df_filter = df.loc[((df["Date"]>=start_date)&(df["Date"]<=end_date))]
plt.plot(df_filter["Date"],df_filter["Open"])
plt.xlabel("Date")
plt.ylabel("Stock value")
plt.title("Stock value vs Day")
plt.show()
```

14. Write a Pandas program to create a plot of adjusted closing prices, thirty 
days and forty days simple moving average of Alphabet Inc. between two specific 
dates.

```python
start_date = pd.to_datetime("2020-04-07")
end_date = pd.to_datetime("2020-09-24")
df["Date"]=pd.to_datetime(df["Date"])
df_filter = df.loc[((df["Date"]>=start_date)&(df["Date"]<=end_date))]
thirty_avg=[]
forty_avg = []
pos = np.arange(len(df_filter))
for i in range(len(df_filter)-30):
    thirty_avg.append(df_filter["Open"][i:i+30].mean())
for i in range(len(df_filter)-40):
    forty_avg.append(df_filter["Open"][i:i+30].mean())
plt.plot(pos,df_filter["Open"],label ="Open")
plt.plot(pos[30:],thirty_avg,label ="Thirty day avg")
plt.plot(pos[40:],forty_avg,label ="Forty day avg")
plt.ylabel("Stock value")
plt.title("Opening stock prices of Alphabet Inc. vs Day")
plt.legend()
plt.xticks(pos[::10],df_filter["Date"][::10],rotation = 45);
plt.show()
```

15. Write a Pandas program to create a plot of adjusted closing prices, 30 days 
simple moving average and exponential moving average of Alphabet Inc. 
between two specific dates.

```python
start_date = pd.to_datetime("2020-04-07")
end_date = pd.to_datetime("2020-09-24")
df["Date"]=pd.to_datetime(df["Date"])
df_filter = df.loc[((df["Date"]>=start_date)&(df["Date"]<=end_date))]
thirty_avg=[]
exp_avg = []
pos = np.arange(len(df_filter))
thirty_avg = df_filter["Close"].rolling(window=30).mean()
exp_avg = df_filter["Close"].ewm(span=30).mean()
plt.plot(pos,df_filter["Open"],label ="Open")
plt.plot(pos,thirty_avg,label ="Thirty day avg")
plt.plot(pos,exp_avg,label ="Forty day avg")
plt.ylabel("Stock value")
plt.title("Opening stock prices of Alphabet Inc. vs Day")
plt.legend()
plt.xticks(pos[::10],df_filter["Date"][::10],rotation = 45);
plt.show()
```

17. Write a Pandas program to create a plot to visualize daily percentage returns 
of Alphabet Inc. stock price between two specific dates.

```python
start_date = pd.to_datetime("2020-04-07")
end_date = pd.to_datetime("2020-09-24")
df["Date"]=pd.to_datetime(df["Date"])
df_filter = df.loc[((df["Date"]>=start_date)&(df["Date"]<=end_date))]
gain = 100*(df_filter["Close"]-df_filter["Open"])/df_filter["Open"]
plt.plot(df_filter["Date"],gain)
plt.xlabel("Date")
plt.ylabel("Gain %")
plt.title("Daily Gain %");
plt.show()
```

18. Write a Pandas program to plot the volatility over a period of time of Alphabet 
Inc. stock price between two specific dates.

```python
start_date = pd.to_datetime("2020-04-07")
end_date = pd.to_datetime("2020-09-24")
df["Date"]=pd.to_datetime(df["Date"])
df_filter = df.loc[((df["Date"]>=start_date)&(df["Date"]<=end_date))]
df_filter = df_filter[['Date', 'Close']]
df_filter.set_index('Date',inplace = True)
volatility = df_filter.asfreq('D', method='ffill').pct_change().rolling(window=30, min_periods=30).std()
volatility.plot()
plt.xlabel("Date")
plt.ylabel("Gain %")
plt.title("Daily Gain %");
plt.show()
```
