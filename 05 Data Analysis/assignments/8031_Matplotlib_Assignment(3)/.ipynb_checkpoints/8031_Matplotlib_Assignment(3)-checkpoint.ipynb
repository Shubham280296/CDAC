5. Write a Python program to plot two or more lines on same plot with suitable
legends of each line.

```python
x = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]

plt.plot(x,y1, label = "line1")
plt.plot(x,y2, label = "line2")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two or more lines on same plot with suitable legends')
plt.legend()
plt.show()
```

6.Write a Python program to plot two or more lines with legends, different
widths and colors.

```python
x = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]

plt.plot(x,y1, label = "line1-width-3", lw = 3, color = "blue")
plt.plot(x,y2, label = "line2-width-5", lw = 5, color = "red")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two or more lines on same plot with suitable legends')
plt.legend()
plt.show()
```

7. Write a Python program to plot two or more lines with different styles.


```python
x = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]
plt.plot(x,y1, label = "line1-dotted", lw = 1, color = "blue", linestyle = ":")
plt.plot(x,y2, label = "line2-dashed", lw = 2, color = "red", linestyle = "--")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two or more lines on same plot with suitable legends')
plt.legend()
plt.show()
```

8. Write a Python program to plot two or more lines and set the line markers.

```python
x = [0,4,5,6,7]
y1 = [2,6,3,6,3]
plt.plot(x,y1, label = "line1-dotted", lw = 1, color = "blue", linestyle = ":",marker='o',markerfacecolor='r')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Display marker")
plt.show()
```

9. Write a Python program to display the current axis limits values and set new
axis values.

```python
x = np.linspace(0,60,120)
y= 2.5*x
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Draw a line")
plt.xlim([0,100])
plt.ylim([0,200])
plt.show()
```

10. Write a Python program to plot quantities which have an x and y position.

```python
x1 = [2,3,5,6,8]
y1 = [1,5,10,18,21]
x2=[3,4,6,7,9]
y2 = [2,6,12,21,22]
plt.scatter(x1,y1,color='b',marker="*")
plt.scatter(x2,y2,color='r',marker=".")
plt.xlim([0,10])
plt.ylim([0,30])
plt.show()
```

11. Write a Python program to plot several lines with different format styles in
one command using arrays.

```python
x = np.linspace(0,5,25)
y1=x
y2=x**2
y3 = x**3
plt.plot(x,y1,color='g', linestyle = "dashed",lw = 1)
plt.scatter(x,y2,color='b',marker='s')
plt.scatter(x,y3,color='r',marker="^")
plt.show()
```

14. Write a Python program to display the grid and draw line charts of the
closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016.
Customized the grid lines with rendering with a larger grid (major grid) and a
smaller grid (minor grid).Turn on the grid but turn off ticks.

```python
x= ['2016-10-03','2016-10-04','2016-10-05','2016-10-06','2016-10-07']
y = [772.559998,776.429993,776.469971,776.859985,775.080017]
plt.plot(x,y,marker='.',color='r')
plt.xlabel("Date")
plt.ylabel("Closing value")
plt.title("Closing stock value of Alphabet Inc.")
plt.grid(which = 'major',color='r',lw=0.5)
plt.minorticks_on()
plt.grid(which='minor',color='grey',ls=':')
```
