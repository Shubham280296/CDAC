```python
import empiricaldist
from statadict import parse_stata_dict
```

```python
dict_file = 'GSS.dct'
data_file = 'GSS.dat.gz'
```

```python
stata_dict = parse_stata_dict(dict_file)
stata_dict.colspecs
```

```python
import gzip
fp = gzip.open(data_file)

import pandas as pd
gss = pd.read_fwf(fp,names=stata_dict.names,clospecs = stata_dict.colspecs)
```

```python
print(gss.shape)
gss.head()
```

```python
print(gss['EDUC'].value_counts().sort_index())
```

```python
import numpy as np
educ = gss['EDUC'].replace([98,99],np.nan)
```

```python
import matplotlib.pyplot as plt
educ.hist(grid = False)
plt.xlabel("Years of education")
plt.ylabel("Number of respondents")
plt.title("Histogram of level of education")
plt.show()
```

```python
from empiricaldist import Pmf
pmf_educ = Pmf.from_seq(educ,normalize=False)
print(type(pmf_educ))
print(pmf_educ.head())
print(pmf_educ.tail())
print(pmf_educ[20])
```

```python
pmf_educ_norm = Pmf.from_seq(educ,normalize=True)
print(pmf_educ_norm.head())
print(pmf_educ_norm[12])
```

```python
pmf_educ_norm.bar(label = 'EDUC')
plt.xlabel('Years od education')
plt.xticks(range(0,21,4))
plt.ylabel('PMF')
plt.title("Distribution of years of education")
plt.legend()
plt.show()
```

```python
from empiricaldist import Cdf
educ = gss['EDUC'].replace([98,99],np.nan)
cdf_educ = Cdf.from_seq(educ)
cdf_educ.plot()
plt.xlabel('Years od education')
plt.xticks(range(0,21,4))
plt.ylabel('CFD')
plt.title("Distribution of years of education")
plt.show()
```

```python
from empiricaldist import Cdf
age = gss['AGE'].replace([98,99],np.nan)
cdf_age = Cdf.from_seq(age)
cdf_age.plot()
plt.xlabel('Age (Years)')
plt.ylabel('CFD')
plt.title("Distribution of age")
plt.show()
```

```python
q = 51
p = cdf_age(q)
p 
# p% people are 51 years old or younger
```

```python
p1 = 0.25
q1 = cdf_age.inverse(p1)
q1
# p1% are of age q1 or less
```

```python
p3 = 0.75
q3 = cdf_age.inverse(p3)
print(q3)
print(q3-q1)
```

```python
print(f"Q1 : {q1},  Q3 : {q3}")
```

```python
Cdf_age = empiricaldist.Cdf.from_seq(age)
Cdf_values = Cdf_age.values
data_points = Cdf_age.index
plt.plot(data_points,Cdf_values)
plt.axvline(x = q1,color="red",linestyle= '--',label = 'Q1')
plt.axvline(x = q3,color="red",linestyle= '--',label = 'Q3')
plt.xlabel('Age (Years)')
plt.ylabel('CFD')
plt.title("Distribution of age")
plt.legend()
plt.show()
```

```python
age_data = gss['AGE']
plt.hist(age_data,bins=20,density =True,alpha = 0.6,color= "b",label = 'PDF')
plt.xlabel('Age (Years)')
plt.ylabel('PDF')
plt.title("PDF of age")
plt.legend()
plt.show()
```
