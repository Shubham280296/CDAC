```python
# KDE kernel density estimation
```

```python
np.random.seed(42)
sample_ages= np.random.randint(0,100,200)
sample_ages = pd.DataFrame(sample_ages,columns=["age"])
sample_ages.head(3)
```

```python
sns.rugplot(data =sample_ages,x = 'age',height = 0.5)
```

```python
sns.displot(data =sample_ages,x = 'age',bins =10,rug = True)
```

```python
sns.displot(data =sample_ages,x = 'age',bins =10,rug = True,kde= True)
```

```python
sns.kdeplot(data=sample_ages,x="age")
```

```python
df = pd.read_csv("country_table.csv")
df
```

```python
df = df.set_index('Countries')
sns.heatmap(df)
plt.show()
```

```python
rates=df.drop('Life expectancy',axis=1)
sns.heatmap(rates)
```

```python
sns.heatmap(rates,linewidth = 0.5)
```

```python
sns.heatmap(rates,linewidth = 0.5,annot = True,cmap = 'viridis')
```

```python
sns.heatmap(rates,linewidth = 0.5,annot = True,cmap = 'viridis',center = 40)
```

```python
sns.heatmap(rates,linewidth = 0.5,annot = True,cmap = 'viridis',center = 1)
```
