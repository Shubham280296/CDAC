```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Basics
```

```python
x = np.arange(0,10)
y = 2*x
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("y=2x")
```

```python
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("y=2x")
plt.xlim(0,6)
plt.ylim(0,12)
#plt.savefig('y=2x.png')
```

```python
x = np.arange(1,10)
y = np.arange(1,19)
y = y[:9]
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("y=x")
plt.scatter(x,y,s=25)
```

```python
x = np.arange(0,10)
y = 2*x
print(x)
print(y)

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
a = np.linspace(0,10,11)
b= a**4
print(a)
print(b)

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.4,0.4])
axes1.plot(a,b)
axes1.set_xlabel("X axis")
axes1.set_ylabel("Y axis")
axes1.set_title("Big fig")
axes2.plot(a,b)
axes2.set_title("Small fig")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.3,0.3])
axes1.plot(a,b)
axes1.set_xlabel("X axis")
axes1.set_ylabel("Y axis")
axes1.set_title("Big fig")
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("Small fig")
plt.show()
```

```python
x = np.arange(0,10)
y = 2*x
print(x)
print(y)

a = np.linspace(0,10,11)
b= a**4
print(a)
print(b)

fig,axes = plt.subplots()
axes.plot(x,y,'b')
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("y=2*x")
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
print(axes)
print(axes.shape)
plt.show()
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)
print(axes)
print(axes.shape)
plt.show()
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2,figsize=(12,8))
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(b,a)
axes[1][0].plot(y,x)
fig.subplots_adjust(left = None,bottom=None,right = None,top=None,wspace = 0.1,hspace=0.1)
plt.show()
```

```python
x= np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x**2')
ax.plot(x,x**3,label='x**3')
ax.legend()
plt.show()
```

```python
x= np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x**2')
ax.plot(x,x**3,label='x**3')
ax.legend(loc = 1)
plt.show()
```

```python
x= np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'b.-',label='x**2')
ax.plot(x,x**3,'g--',label='x**3')
ax.legend(loc = 1)
plt.show()
```

```python
x= np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,color = 'blue',alpha = 0.5,label='x**2')
ax.plot(x,x**3,color = 'green',alpha = 0.1,label='x**3')
ax.legend(loc = 1)
plt.show()
```

```python
x= np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x+2,color = 'red',linewidth = 0.25,label='x+2')
ax.plot(x,x+3,color = 'red',lw =0.5,label='x+3')
ax.plot(x,x+1,color = 'red',linewidth = 1,label='x+1')
ax.plot(x,x+4,color = 'red',lw =10,label='x+4')
ax.legend(loc = 1)
plt.show()
```

```python
x= np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x+2,color = 'red',linewidth = 0.25,label='x+2',linestyle='-')
ax.plot(x,x+3,color = 'red',lw =0.5,label='x+3',ls='-.')
ax.plot(x,x+1,color = 'red',linewidth = 1,label='x+1',ls=':')
ax.plot(x,x+4,color = 'red',lw =10,label='x+4',ls = '--')
ax.legend(loc = 1)
plt.show()
```

```python
x= np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x+2,marker= '+',markersize = 10)
ax.plot(x,x+3,marker='o',ms=10)
ax.plot(x,x+1,marker='*',ms=10,lw= 0)
ax.plot(x,x+4,marker='1',ms=10)
plt.show()
```
