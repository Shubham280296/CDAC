```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
country = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Unsupervised Learning on Country Data\Country-data.csv",index_col=0)
country.head()
```

```python
############ AgglomerativeClustering
```

```python
scaler = StandardScaler().set_output(transform='pandas')
country_scaled = scaler.fit_transform(country)

link = "average"
mergings = linkage(country_scaled,method=link)
dendrogram(mergings,
           labels=list(country_scaled.index))
plt.title(link+" linkage")
plt.show()
```

```python
clust = AgglomerativeClustering(n_clusters=2, linkage='ward')
clust.fit(country_scaled)

country_clust = country.copy()
country_clust['Clust'] = clust.labels_.astype(str)
print(silhouette_score(country_scaled, clust.labels_))
country_clust.head()
```

```python
Ks = np.arange(2,13)
scores = []
for i in Ks:
    clust = AgglomerativeClustering(n_clusters=i, linkage='ward')
    clust.fit(country_scaled)
    scores.append(silhouette_score(country_scaled, clust.labels_))

i_max = np.argmax(scores)
print("Best no. of clusters:", Ks[i_max])
print("Best Score:", scores[i_max])
```

```python
clust = AgglomerativeClustering(n_clusters=2, linkage='ward')
clust.fit(country_scaled)

prcomp = PCA().set_output(transform='pandas')
principalComponents = prcomp.fit_transform(country_scaled)
print(np.cumsum(prcomp.explained_variance_ratio_*100))

principalComponents["Clust"]=clust.labels_.astype(str)
sns.scatterplot(x='pca0',y='pca1',data=principalComponents, hue='Clust')
#for i in np.arange(0,country_scaled.shape[0]):
#    plt.text(principalComponents.values[i,0],principalComponents.values[i,1],list(country_scaled.index)[i])
plt.show()
```

```python
######### Kmeans
```

```python
scaler = StandardScaler().set_output(transform='pandas')
country_scaled = scaler.fit_transform(country)

Ks = np.arange(2,13)
scores = []
for i in Ks:
    clust = KMeans(n_clusters=i,random_state=24,init='random')
    clust.fit(country_scaled)
    scores.append(silhouette_score(country_scaled, clust.labels_))

i_max = np.argmax(scores)
print("Best no. of clusters:", Ks[i_max])
print("Best Score:", scores[i_max])
```

```python
clust = KMeans(n_clusters= Ks[i_max],random_state=24,init='random')
clust.fit(country_scaled)

prcomp = PCA().set_output(transform='pandas')
principalComponents = prcomp.fit_transform(country_scaled)
print(np.cumsum(prcomp.explained_variance_ratio_*100))

principalComponents["Clust"]=clust.labels_.astype(str)
sns.scatterplot(x='pca0',y='pca1',data=principalComponents, hue='Clust')
#for i in np.arange(0,country_scaled.shape[0]):
#    plt.text(principalComponents.values[i,0],principalComponents.values[i,1],list(country_scaled.index)[i])
plt.show()
```

```python
clust_country = country.copy()
clust_country['Clust'] = clust.labels_.astype(str)

clust_country.groupby('Clust').mean()
```

```python
g = sns.FacetGrid(clust_country,col='Clust')
g.map(plt.hist,'child_mort')
plt.show()
```

```python
g = sns.FacetGrid(clust_country,col='Clust')
g.map(sns.boxplot,'health')
plt.show()
```

```python
Ks = np.arange(2,25)
inertia = []
for i in Ks:
    clust = KMeans(n_clusters=i,random_state=24,init='random')
    clust.fit(country_scaled)
    inertia.append(clust.inertia_)
```

```python
plt.plot(Ks,inertia,marker='*')
plt.title("WSS vs number of clusters(Scree plot)")
plt.xlabel("No of clusters")
plt.ylabel("WSS");
plt.show()
```
