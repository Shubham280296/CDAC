```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
```

```python
train = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Assignments\Flood Prediction\train.csv")
test = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Assignments\Flood Prediction\test.csv")
```

```python
X_train = train.drop(["id","FloodProbability"],axis=1)
X_test = test.drop("id",axis=1)
y_train = train["FloodProbability"]
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
submit = pd.DataFrame({"id":test["id"],"FloodProbability":y_pred})
submit
```

```python
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Assignments\Flood Prediction\submit_lr.csv",index=False)
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)
elastic = ElasticNet()
params= {"alpha": np.linspace(0.0001,4,40),"l1_ratio":np.linspace(0.001,1,10)}
gcv = GridSearchCV(elastic,param_grid=params,cv=kfold)# default scoring r2

gcv.fit(X_train,y_train)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
pd.DataFrame(gcv.cv_results_)
```

```python
y_pred= gcv.best_estimator_.predict(X_test)
```

```python
submit = pd.DataFrame({"id":test["id"],"FloodProbability":y_pred}).round(decimals=3)
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Assignments\Flood Prediction\submit_el.csv",index=False)
```
