```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import  CatBoostRegressor
```

```python
train = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Flood Prediction\train.csv")
test = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Flood Prediction\test.csv")
```

```python
X = train.drop(["id","FloodProbability"],axis=1)
X_test = test.drop("id",axis=1)
y = train["FloodProbability"]
```

```python
xgb = XGBRegressor(random_state=24)
xgb.fit(X,y)
y_pred= xgb.predict(X_test)
submit = pd.DataFrame({"id":test["id"],"FloodProbability":y_pred}).round(decimals=3)
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Flood Prediction\xgb.csv",index=False)
```

```python
lgbm = LGBMRegressor(random_state=24)
lgbm.fit(X,y)
y_pred= lgbm.predict(X_test)
submit = pd.DataFrame({"id":test["id"],"FloodProbability":y_pred}).round(decimals=3)
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Flood Prediction\lgbm.csv",index=False)
```

```python
cate = list(X.select_dtypes(include=object).columns)
cat =  CatBoostRegressor(random_state=24,cat_features=cate)
cat.fit(X,y)
y_pred= cat.predict(X_test)
submit = pd.DataFrame({"id":test["id"],"FloodProbability":y_pred}).round(decimals=3)
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Flood Prediction\cat.csv",index=False)
```
