```python
item_weight_tr = train.iloc[:,0:2]
item_weight_tst = test.iloc[:,0:2]

item_weight = pd.concat([item_weight_tst,item_weight_tr])

item_weight = item_weight.drop_duplicates()
item_weight = item_weight.dropna()

train.drop('Item_Weight',axis=1,inplace= True)
test.drop('Item_Weight',axis=1,inplace= True)

train = train.merge(item_weight,how="left",on='Item_Identifier')
test = test.merge(item_weight,how="left",on='Item_Identifier')
train.info()
```

```python
outlet_tr = train.iloc[:,5:10]
outlet_tst = test.iloc[:,5:10]

outlet = pd.concat([outlet_tr,outlet_tst])

outlet = outlet.drop_duplicates()
outlet['Outlet_Size'].fillna(value='small',inplace=True)

train.drop('Outlet_Size',axis=1,inplace= True)
test.drop('Outlet_Size',axis=1,inplace= True)

train = train.merge(outlet,how="left",on=['Outlet_Identifier','Outlet_Establishment_Year','Outlet_Location_Type','Outlet_Type'])
test = test.merge(outlet,how="left",on=['Outlet_Identifier','Outlet_Establishment_Year','Outlet_Location_Type','Outlet_Type'])
train.info()
```

```python
train.drop(['Item_Identifier','Outlet_Identifier'],inplace=True,axis=1)
X_test = test.drop(['Item_Identifier','Outlet_Identifier'],axis=1)

X_train = train.drop(['Item_Outlet_Sales'],axis=1)
y_train = train['Item_Outlet_Sales']
```

```python
X_train = pd.get_dummies(X_train,drop_first=True)
X_test = pd.get_dummies(X_test,drop_first=True)
```

```python
elastic = ElasticNet()
lr = LinearRegression()
xgb = XGBRegressor(random_state=24)
lgbm = LGBMRegressor(random_state=24)
cat =  CatBoostRegressor(random_state=24)
stack3 = StackingRegressor([('LGBM',lgbm),('Elastic',elastic),('XGB',xgb),('LR',lr)],final_estimator=cat,passthrough=True )

stack3.fit(X_train,y_train)

y_pred  = stack3.predict(X_test)
submit = pd.DataFrame({"Item_Identifier":test["Item_Identifier"],"Outlet_Identifier":test["Outlet_Identifier"],"Item_Outlet_Sales":y_pred})
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Big mart sales practice problem (AnalyticsVidya)\stack3.csv",index=False)
```

```python
clf = IsolationForest(contamination=0.05, random_state=24)
clf.fit(X_train)
predictions = clf.predict(X_train)

print("%age of outliers="+ str((predictions<0).mean()*100)+ "%")
abn_ind = np.where(predictions < 0)
print("Outliers:")
outliers = X_train.index[abn_ind]
print(outliers)
```
