```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import StackingClassifier
```

```python
df = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Cirrhosis\train.csv")
X = pd.get_dummies(df.drop(["Status","id"],axis=1),drop_first=True)
le = LabelEncoder()
y = le.fit_transform(df["Status"])
```

```python
test = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Cirrhosis\test.csv")
X_test = pd.get_dummies(test.drop("id",axis=1),drop_first=True)
```

```python
#### Stack 1 linear,lgbm,xgb ---> cat
```

```python
lr = LogisticRegression()
xgb = XGBClassifier(random_state=24)
lgbm = LGBMClassifier(random_state=24)
cat = CatBoostClassifier(random_state =24)
stack1 = StackingClassifier([('LR',lr),('LGBM',lgbm),('XGB',xgb)],final_estimator=cat,passthrough=True)

stack1.fit(X,y)

y_pred_prob = stack1.predict_proba(X_test)
submit = pd.DataFrame({"id":test["id"],"Status_C":y_pred_prob[:,0],"Status_CL":y_pred_prob[:,1],"Status_D":y_pred_prob[:,2]})
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Cirrhosis\stack1.csv",index=False)
```

```python
#### Stack 2 lgmb,linear,cat ---> linear
```

```python
lr = LogisticRegression()
xgb = XGBClassifier(random_state=24)
cat =  CatBoostClassifier(random_state=24)
lgbm = LGBMClassifier(random_state=24)
stack2 = StackingClassifier([('LGBM',lgbm),('XGB',xgb),('CAT',cat)],final_estimator=lr,passthrough=True )

stack2.fit(X,y)

y_pred_prob = stack2.predict_proba(X_test)
submit = pd.DataFrame({"id":test["id"],"Status_C":y_pred_prob[:,0],"Status_CL":y_pred_prob[:,1],"Status_D":y_pred_prob[:,2]})
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Cirrhosis\stack2.csv",index=False)
```
