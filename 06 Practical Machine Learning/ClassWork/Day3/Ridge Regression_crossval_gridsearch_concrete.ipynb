```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,cross_val_score, KFold
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Cases\Concrete Strength\Concrete_Data.csv")
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.3, random_state=24)

rd = Ridge(alpha = 100)
rd.fit(X_train, y_train)

y_pred = rd.predict(X_test)

############## Model Evaluation ##############
print(r2_score(y_test, y_pred))
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)
lambdas = np.linspace(0,100,500)
r2_scores=[]
for i in lambdas:
    rd = Ridge(alpha = i)
    results = cross_val_score(rd,X, y,cv = kfold,scoring = 'r2')#scoring is r2 default
    r2_scores.append(results.mean())

lambdas[np.array(r2_scores).argmax()]
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)
ridge = Ridge()
from sklearn.model_selection import GridSearchCV
params= {"alpha": np.linspace(0,100,500)}
gcv = GridSearchCV(ridge,param_grid=params,cv=kfold)# default scoring r2

gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
pd.DataFrame(gcv.cv_results_)
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)
ridge = Ridge()
from sklearn.model_selection import GridSearchCV
params= {"alpha": np.linspace(0,100,500)}
gcv = GridSearchCV(ridge,param_grid=params,cv=kfold,scoring="neg_mean_squared_error")#maximumizes negative of mean squared error

gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
pd.DataFrame(gcv.cv_results_)
```
