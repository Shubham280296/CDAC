```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
housing = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Datasets\Housing.csv")
housing.head()
```

```python
price_int = [(0,50000),(50000,100000),(100000,150000),(150000,200000)]
bins_price = pd.IntervalIndex.from_tuples(price_int)
housing['price_slab'] = pd.cut(housing['price'],bins_price)

area_int = [(0,5000),(5000,10000),(10000,15000),(15000,20000)]
bins_area = pd.IntervalIndex.from_tuples(area_int)
housing['area_slab'] = pd.cut(housing['lotsize'],bins_area)

housing.drop(['price','lotsize'],axis=1,inplace=True)

housing = housing.astype(object)
housing_dum = pd.get_dummies(housing,prefix_sep='_')
```

```python
housing_dum.head()
```

```python
itemsets = apriori(housing_dum, min_support = 0.5,use_colnames=True)

rules = association_rules(itemsets, metric= 'confidence', min_threshold = 0.7)

rules[['antecedents','consequents','support','confidence','lift']]
```

```python
rules.sort_values(by = 'lift', ascending=False).head(6)
```

```python
rules_df = rules.sort_values(by = ['lift','confidence'], ascending=False)
relv_rules = rules_df[rules_df['lift']>1]
relv_rules
```
