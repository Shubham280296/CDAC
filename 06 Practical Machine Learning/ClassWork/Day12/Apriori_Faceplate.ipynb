```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Datasets\Faceplate.csv",index_col=0)
df.head()
```

```python
df = df.astype(bool)
itemsets = apriori(df, min_support = 0.2,use_colnames=True)

rules = association_rules(itemsets, metric= 'confidence', min_threshold = 0.6)

print(rules[['antecedents','consequents','support','confidence','lift']])
```

```python
rules.sort_values(by = 'lift', ascending=False).head(6)
```

```python
rules_df = rules.sort_values(by = ['lift','confidence'], ascending=False)
relv_rules = rules_df[rules_df['lift']>1]
relv_rules
```
