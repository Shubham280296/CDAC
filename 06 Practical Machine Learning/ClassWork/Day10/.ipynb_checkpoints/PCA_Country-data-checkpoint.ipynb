```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from pca import pca
```

```python
country = pd.read_csv(r'C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Unsupervised Learning on Country Data\Country-data.csv',index_col=0)
country
```

```python
scaler = StandardScaler().set_output(transform='pandas')
country_scaled=scaler.fit_transform(country)
```

```python
prcomp = PCA().set_output(transform='pandas')
principalComponents = prcomp.fit_transform(country_scaled)
principalComponents.head()
```

```python
print(np.cumsum(prcomp.explained_variance_))
print(np.cumsum(prcomp.explained_variance_ratio_))
print(np.cumsum(prcomp.explained_variance_ratio_*100))
print(np.cumsum(prcomp.explained_variance_ratio_*100)[6])
```

```python
ys = np.cumsum(prcomp.explained_variance_ratio_*100)
xs = np.arange(1,10)
plt.plot(xs,ys)
```

```python
model = pca()
results = model.fit_transform(country_scaled, col_labels=country_scaled.columns,row_labels =list(country_scaled.index) )
model.biplot(label=True, legend=True)
for i in np.arange(0,country_scaled.shape[0]):
    plt.text(principalComponents.values[i,0],principalComponents.values[i,1],list(country_scaled.index)[i])
plt.show()
```
