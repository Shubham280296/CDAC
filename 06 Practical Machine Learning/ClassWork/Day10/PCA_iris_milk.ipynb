```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from pca import pca
```

```python
iris = pd.read_csv(r'C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Datasets\iris.csv')
```

```python
iris
```

```python
scaler = StandardScaler().set_output(transform='pandas')
scaler.fit(iris.drop("Species",axis=1))
irisscaled = scaler.transform(iris.drop("Species",axis=1))
```

```python
prcomp = PCA().set_output(transform='pandas')
principalComponents = prcomp.fit_transform(irisscaled)
principalComponents.head()
```

```python
print(np.cumsum(prcomp.explained_variance_))
print(np.cumsum(prcomp.explained_variance_ratio_))
print(np.cumsum(prcomp.explained_variance_ratio_*100))
print(np.cumsum(prcomp.explained_variance_ratio_*100)[1])
```

```python
ys = np.cumsum(prcomp.explained_variance_ratio_*100)
xs = np.arange(1,5)
plt.plot(xs,ys)
```

```python
principalComponents["Species"]=iris["Species"]
sns.scatterplot(x='pca0',y='pca1',data=principalComponents, hue='Species')
plt.show()
```

```python
sns.pairplot(data=principalComponents,hue='Species')
plt.show()
```

```python
principalComponents["Species"][1]
```

```python
model = pca()
results = model.fit_transform(irisscaled, col_labels=irisscaled.columns,row_labels =iris["Species"] )
model.biplot(label=True, legend=True)
```
