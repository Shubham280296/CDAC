```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,cross_val_score, KFold
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Cases\Concrete Strength\Concrete_Data.csv")
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)
lasso = Lasso()
params= {"alpha": np.linspace(0.0001,100,4)}
gcv = GridSearchCV(lasso,param_grid=params,cv=kfold)# default scoring r2

gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)
pd.DataFrame(gcv.cv_results_)
```
