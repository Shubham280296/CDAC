```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
# Boston
```

```python
boston = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Datasets\Boston.csv")
X = boston.drop("medv",axis=1)
y = boston["medv"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=24)

poly = PolynomialFeatures(degree=2).set_output(transform='pandas')
lr = LinearRegression()
pipe = Pipeline([("Poly",poly),("LR",lr)])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
```

```python
# Pipeline in cross validation 
```

```python
boston = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Datasets\Boston.csv")
X = boston.drop("medv",axis=1)
y = boston["medv"]
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)
lr = LinearRegression()
degrees= [1,2,3,4,5]
scores = []

for i in degrees:
    poly = PolynomialFeatures(degree=i)
    pipe = Pipeline([("Poly",poly),("LR",lr)])
    results = cross_val_score(pipe,X, y,cv = kfold)
    scores.append(results.mean())
```

```python
i_max = np.argmax(scores)
print("Best degree - ",degrees[i_max])
print("Best score - ",scores[i_max])
```

```python
# Pipeline in grid search
```

```python
boston = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Datasets\Boston.csv")
X = boston.drop("medv",axis=1)
y = boston["medv"]

kfold = KFold(n_splits =5,shuffle = True, random_state = 24)

lr = LinearRegression()
poly = PolynomialFeatures()
pipe = Pipeline([("Poly",poly),("LR",lr)])

print(pipe.get_params())
```

```python
params = {"Poly__degree":[1,2,3,4,5]}
gcv = GridSearchCV(pipe,param_grid = params,cv=kfold)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
boston = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Datasets\Boston.csv")
X = boston.drop("medv",axis=1)
y = boston["medv"]

kfold = KFold(n_splits =5,shuffle = True, random_state = 24)

ridge = Ridge()
poly = PolynomialFeatures()
pipe = Pipeline([("Poly",poly),("Ridge",ridge)])

print(pipe.get_params())
```

```python
# Pipeline in grid search using lasso
```

```python
boston = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Datasets\Boston.csv")
X = boston.drop("medv",axis=1)
y = boston["medv"]

kfold = KFold(n_splits =5,shuffle = True, random_state = 24)

lasso = Lasso()
poly = PolynomialFeatures()
pipe = Pipeline([("Poly",poly),("Lasso",lasso)])

print(pipe.get_params())

params = {"Poly__degree":[1,2,3,4,5],'Lasso__alpha':np.linspace(0.001,5,10)}
gcv_lasso = GridSearchCV(pipe,param_grid = params,cv=kfold)
gcv_lasso.fit(X,y)
```

```python
print(gcv_lasso.best_params_)
print(gcv_lasso.best_score_)
```

```python
# Pipeline in grid search using elasticnet
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)

elastic = ElasticNet()
poly = PolynomialFeatures()
pipe = Pipeline([("Poly",poly),("Elastic",elastic)])

print(pipe.get_params())

params = {"Poly__degree":[1,2,3],'Elastic__alpha':np.linspace(0.001,5,10),"Elastic__l1_ratio":np.linspace(0,1,5)}
gcv_elastic = GridSearchCV(pipe,param_grid = params,cv=kfold)
gcv_elastic.fit(X,y)
```

```python
print(gcv_elastic.best_params_)
print(gcv_elastic.best_score_)
```

```python
best_model = gcv_elastic.best_estimator_
print(best_model.named_steps.Elastic.intercept_)
print(best_model.named_steps.Elastic.coef_)
```
