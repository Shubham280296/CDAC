```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

```python
#### milk
```

```python
milk = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Datasets\milk.csv",index_col=0)
milk.head()
```

```python
scaler = StandardScaler().set_output(transform='pandas')
milk_scaled = scaler.fit_transform(milk)
link = "ward"
mergings = linkage(milk_scaled,method=link)
dendrogram(mergings,
           labels=list(milk_scaled.index))
plt.title(link+" linkage")
plt.show()
```

```python
clust = AgglomerativeClustering(n_clusters=2)
clust.fit(milk_scaled)

milk_clust = milk.copy()
milk_clust['Clust'] = clust.labels_.astype(str)
print(silhouette_score(milk_scaled, clust.labels_))
milk_clust.head()
```

```python
Ks = np.arange(2,13)
scores = []
for i in Ks:
    clust = AgglomerativeClustering(n_clusters=i)
    clust.fit(milk_scaled)
    scores.append(silhouette_score(milk_scaled, clust.labels_))

i_max = np.argmax(scores)
print("Best no. of clusters:", Ks[i_max])
print("Best Score:", scores[i_max])
```

```python
clust = AgglomerativeClustering(n_clusters=3)
clust.fit(milk_scaled)

prcomp = PCA().set_output(transform='pandas')
principalComponents = prcomp.fit_transform(milk_scaled)
print(np.cumsum(prcomp.explained_variance_ratio_*100))

principalComponents["Clust"]=clust.labels_.astype(str)
sns.scatterplot(x='pca0',y='pca1',data=principalComponents, hue='Clust')
for i in np.arange(0,milk_scaled.shape[0]):
    plt.text(principalComponents.values[i,0],principalComponents.values[i,1],list(milk_scaled.index)[i])
plt.show()
```
