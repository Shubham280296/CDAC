```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
Concrete = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Cases\Concrete Strength\Concrete_Data.csv")
X = Concrete.drop("Strength",axis=1)
y = Concrete["Strength"]
```

```python
# Pipeline in grid search for LinearRegression
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)

lr = LinearRegression()
poly = PolynomialFeatures()
pipe = Pipeline([("Poly",poly),("LR",lr)])

print(pipe.get_params())

params = {"Poly__degree":[1,2,3,4,5]}
gcv = GridSearchCV(pipe,param_grid = params,cv=kfold)
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
# Pipeline in grid search using ridge
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)

ridge = Ridge()
poly = PolynomialFeatures()
pipe = Pipeline([("Poly",poly),("Ridge",ridge)])

print(pipe.get_params())

params = {"Poly__degree":[1,2,3],'Ridge__alpha':np.linspace(0.001,5,10)}
gcv_ridge= GridSearchCV(pipe,param_grid = params,cv=kfold)
gcv_ridge.fit(X,y)
```

```python
print(gcv_ridge.best_params_)
print(gcv_ridge.best_score_)
```

```python
# Pipeline in grid search using lasso
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)

lasso = Lasso()
poly = PolynomialFeatures()
pipe = Pipeline([("Poly",poly),("Lasso",lasso)])

print(pipe.get_params())

params = {"Poly__degree":[1,2,3],'Lasso__alpha':np.linspace(0.001,5,10)}
gcv_lasso = GridSearchCV(pipe,param_grid = params,cv=kfold)
gcv_lasso.fit(X,y)
```

```python
print(gcv_lasso.best_params_)
print(gcv_lasso.best_score_)
```

```python
# Pipeline in grid search using Elasticnet
```

```python
kfold = KFold(n_splits =5,shuffle = True, random_state = 24)

elastic = ElasticNet()
poly = PolynomialFeatures()
pipe = Pipeline([("Poly",poly),("Elastic",elastic)])

print(pipe.get_params())

params = {"Poly__degree":[1,2,3],'Elastic__alpha':np.linspace(0.001,5,10),"Elastic__l1_ratio":np.linspace(0,1,5)}
gcv_elastic = GridSearchCV(pipe,param_grid = params,cv=kfold)
gcv_elastic.fit(X,y)
```

```python
print(gcv_elastic.best_params_)
print(gcv_elastic.best_score_)
```

Hence Best model obsereved is Ridge

```python
best_model = gcv_ridge.best_estimator_
print(best_model.named_steps.Ridge.intercept_)
print(best_model.named_steps.Ridge.coef_)
```
