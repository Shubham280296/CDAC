```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder, StandardScaler,MinMaxScaler
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
from sklearn.metrics import accuracy_score, log_loss
from sklearn.pipeline import Pipeline
```

```python
kyp = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\ClassWork\Cases\Kyphosis\Kyphosis.csv")
le = LabelEncoder()
y = le.fit_transform(kyp["Kyphosis"])
X = kyp.drop('Kyphosis',axis=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=24,stratify=y)
```

```python
svm = SVC(C =  0.1,kernel = "linear")
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
###### gridsearch
```

```python
svm = SVC(C= 0.1, kernel = "linear")
params = {"C":np.arange(0.1,5,0.1)}
kfold = StratifiedKFold(n_splits =5,shuffle = True, random_state = 24)
gcv = GridSearchCV(svm,param_grid = params,cv=kfold)
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
pd.DataFrame(gcv.cv_results_)
```

```python
### using predict.proba enable probability in parameters for SVC
```

```python
svm = SVC(C =  0.1,kernel = "linear",probability =True, random_state=24)
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
print(accuracy_score(y_test,y_pred))
y_pred_proba = svm.predict_proba(X_test)
print(log_loss(y_test,y_pred_proba))
```

```python
### using predict.proba enable probability in parameters for SVC gridsearch
```

```python
svm = SVC(C =  0.1,kernel = "linear",probability =True, random_state=24)
params = {"C":np.linspace(0.1,5,10)}
kfold = StratifiedKFold(n_splits =5,shuffle = True, random_state = 24)
gcv = GridSearchCV(svm,param_grid = params,cv=kfold,scoring="neg_log_loss")
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
pd.DataFrame(gcv.cv_results_)
```

```python
##### scaling data and using predict.proba enable probability in parameters for SVC gridsearch
```

```python
svm = SVC(C =  0.1,kernel = "linear",probability =True, random_state=24)
scl_mm = MinMaxScaler()
scl_std = StandardScaler()
pipe = Pipeline([("Scl",None),("SVM",svm)])

print(pipe.get_params())
```

```python
params = {"SVM__C":np.linspace(0.001,5,500),"Scl":[scl_mm,scl_std,None]}
kfold = StratifiedKFold(n_splits =5,shuffle = True, random_state = 24)
gcv = GridSearchCV(pipe,param_grid = params,cv=kfold, scoring="neg_log_loss")
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
pd.DataFrame(gcv.cv_results_)
```

```python
############ polynomial kernel scaling data and using predict.proba enable probability in parameters for SVC gridsearch
```

```python
svm = SVC(C =  0.1,kernel = "poly",probability =True, random_state=24)
scl_mm = MinMaxScaler()
scl_std = StandardScaler()
pipe = Pipeline([("Scl",None),("SVM",svm)])

print(pipe.get_params())
```

```python
params = {"SVM__C":np.linspace(0.001,5,20),"Scl":[scl_mm,scl_std,None],"SVM__degree":[2,3],"SVM__coef0":np.linspace(0,3,5)}
kfold = StratifiedKFold(n_splits =5,shuffle = True, random_state = 24)
gcv = GridSearchCV(pipe,param_grid = params,cv=kfold, scoring="neg_log_loss",verbose=2)
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
pd.DataFrame(gcv.cv_results_)
```

```python
svm = SVC(C =  0.1,kernel = "rbf",probability =True, random_state=24)
scl_mm = MinMaxScaler()
scl_std = StandardScaler()
pipe = Pipeline([("Scl",None),("SVM",svm)])
params = {"SVM__C":np.linspace(0.001,5,20),"Scl":[scl_mm,scl_std,None],"SVM__gamma":np.linspace(0.001,5,5)}
kfold = StratifiedKFold(n_splits =5,shuffle = True, random_state = 24)
gcv = GridSearchCV(pipe,param_grid = params,cv=kfold, scoring="neg_log_loss",verbose=2)
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```
