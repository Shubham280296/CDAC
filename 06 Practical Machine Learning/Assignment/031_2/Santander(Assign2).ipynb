Q) Consider the dataset at the link Santander Customer Satisfaction | Kaggle. 

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
train = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Santander Customer Satisfaction(Assign2)\train.csv",index_col=0)
test = pd.read_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Santander Customer Satisfaction(Assign2)\test.csv",index_col=0)
```

```python
X_train = train.drop("TARGET",axis=1)
y_train = train["TARGET"]
```

```python
## A) Find out how many principal components explain more than 90% variation taking all the variables except ID and target. 
```

```python
scaler = MinMaxScaler().set_output(transform='pandas')
scaler.fit(X_train)
X_scl_trn = scaler.transform(X_train)

pca = PCA().set_output(transform='pandas')
pca_x_trn = pca.fit_transform(X_scl_trn)
cumsum = pd.Series(np.cumsum(pca.explained_variance_ratio_*100))
print(cumsum)

pca_90 = (cumsum<90).sum()+1
print("Principal components needed to explain more than 90% variation",pca_90)
```

From principal components cumulative sum, it is observed that first 12 principal components explain more than 90% variation

```python
#### B) Try the following models with PCA transform (Pipeline):
######## Random Forest ###########
```

```python
scaler = MinMaxScaler().set_output(transform='pandas')
prcomp = PCA(n_components=pca_90).set_output(transform='pandas')
rf = RandomForestClassifier(random_state=24)

pipe_rf = Pipeline([("SCL",scaler),("PCA",prcomp),("TREE",rf)])

pipe_rf.fit(X_train,y_train)

y_pred_prob  = pipe_rf.predict_proba(test)[:,1]
submit = pd.DataFrame({"ID":test.index,"TARGET":y_pred_prob})
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Santander Customer Satisfaction(Assign2)\RandomForest.csv",index=False)
```

```python
######## XGBoost ##########
```

```python
scaler = MinMaxScaler().set_output(transform='pandas')
prcomp = PCA(n_components=pca_90).set_output(transform='pandas')
xgb = XGBClassifier(random_state=24)

pipe_xgb = Pipeline([("SCL",scaler),("PCA",prcomp),("XGB",xgb)])

pipe_xgb.fit(X_train,y_train)

y_pred_prob  = pipe_xgb.predict_proba(test)[:,1]
submit = pd.DataFrame({"ID":test.index,"TARGET":y_pred_prob})
submit.to_csv(r"C:\Users\Administrator.DAI-PC2\Downloads\Shubham\Practical Machine Learning\Kaggle\Santander Customer Satisfaction(Assign2)\XGB.csv",index=False)
```
